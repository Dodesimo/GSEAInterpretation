```python
adata_copy = adata.copy(filename='CRC_CD45.cms.h5ad')
def get_cluster_proportions(adata, cluster_key="Cell_subtype", sample_key="replicate", drop_values=None):
#    adata_tmp = DC_adata_copy.to_memory()
    adata_tmp = adata_copy
 
    sizes = adata_tmp.obs.groupby([sample_key, cluster_key]).size().reset_index(name='size')
    props = sizes.pivot_table(index=sample_key, columns=cluster_key, values='size', fill_value=0)
    props = props.div(props.sum(axis=1), axis=0) * 100
    if drop_values is not None:
        props = props[~props.index.isin(drop_values)]
    
    return props

a=get_cluster_proportions(adata,
                            cluster_key="Cell_subtype",
                            sample_key="group",
                            drop_values=None)

a
b=a.T
b
```
